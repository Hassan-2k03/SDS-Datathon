Question 6:

Check for any outliers in all numeric columns and then analyze carefully, 
how they should be addressed.

Outliers can be handled by :

1. Replaceing them with central tendencies. 
2. Removing the outliers in case of extreme out of range. 
3. Transforming the data by any of the methods such as dividing the number by larger number or applying log.  

```python
# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number])

# Calculate IQR for each numeric column
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1

# Define a threshold for outliers
threshold = 1.5

# Identify outliers
outliers = ((numeric_cols < (Q1 - threshold * IQR)) | (numeric_cols > (Q3 + threshold * IQR))).any(axis=1)

# Print outliers
print(df[outliers])

# Remove outliers
df = df[~outliers]
print(df)
```

Question 8:

Draw bar plots, visualize and also indicate any insights can be obtained 
by taking  X-axis vs Y-axis as: 

• Cut vs no.of diamonds
